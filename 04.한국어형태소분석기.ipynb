```python
from konlpy.tag import Kkma, Komoran, Okt, Hannanum #Mecab(윈도우 작동 불가능)

okt = Okt()
kkma = Kkma()
komoran = Komoran()
hannanum = Hannanum()

text = '나랏말이 중국과 달라 한자와 서로 통하지 아니하므로, \
    우매한 백성들이 말하고 싶은 것이 있어도 마침내 제 뜻을 잘 표현하지 못하는 사람이 많다.\
    내 이를 딱하게 여기어 새로 스물여덟 자를 만들었으니, \
    사람들로 하여금 쉬 익히어 날마다 쓰는 데 편하게 할 뿐이다.'
```

```python
# .morphs()함수 : 텍스트를 형태소 단위로 처리
print("[Kkma]")
print(kkma.morphs(text))
print("[Okt]")
print(okt.morphs(text))
print("[Komoran]")
print(komoran.morphs(text))
print("[Hannanum]")
print(hannanum.morphs(text))
```

```python
# 정규화(norm), 어간추출(stem)
# stem: 단어에서 어간추출
print(okt.morphs(text, stem= True))
```

```python
# .nouns()함수: 명사 추출
print("[Kkma]")
print(kkma.nouns(text))
print("[Okt]")
print(okt.nouns(text))
print("[Komoran]")
print(komoran.nouns(text))
print("[Hannanum]")
print(hannanum.nouns(text))
```

```python
# .phrases()함수: 어절 추출
print(okt.phrases(text))
```

```python
# .pos()함수: 품사 태깅
print("[Kkma]")
print(kkma.pos(text)) 
print("[Okt]")
print(okt.pos(text))
print("[Komoran]")
print(komoran.pos(text))
print("[Hannanum]")
print(hannanum.pos(text))
```
