```python
import spacy

nlp = spacy.load('en_core_web_sm')

text = 'Wikipedia is maintained by colunteers.'
doc = nlp(text)
```

```python
for token in doc:
    print(token.text,
         token.lemma_, # 표제어
         token.pos_, # 단어의 품사
         token.tag_, # 자세한 품사
         token.dep_, # 문법적 의존 관계
         token.is_stop)# 불용어 여부
```

```python
spacy.explain('PROPN')
```

## 명사와 동사의 표제어로 단어 문서 행렬 만들기

```python
def extract_nv(text):
    doc = nlp(text)
    words = []
    for token in doc:
        if token.tag_[0] in 'NV':
            words.append(token.lemma_.lower())
    return words
```

```python
extract_nv('Apple is a company')
```
