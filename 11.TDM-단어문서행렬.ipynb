```python
import wget

wget.download('https://raw.githubusercontent.com/euphoris/datasets/master/imdb.xlsx')
```

```python
import pandas as pd

df = pd.read_excel('imdb.xlsx',index_col=0)
df.head()
```

```python
df.shape
```

## TDM 만들기

```python
# 토큰이 문서별로 몇 번 등장했는지 행렬로 정리(빈칸 중심)
# max_features : 단어문서행렬에 포함시킬 최대(max)의 단어(feature) 수
# 빈도 순으로 최대 500 단어까지 포함
# stop_words : 불용어 설정(영어의 경우 관사, 전치사 등을 제외)
# 다른 언어는 리스트 등의 스타일로 불용어를 넘겨줘야 함
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=500,stop_words='english')
tdm = cv.fit_transform(df['review'])

tdm
```

```python
tdm.shape
```

## 단어 목록

```python
cv.get_feature_names_out()[:10]
```

```python
len(cv.get_feature_names_out())
```

## 단어별 총 빈도

```python
tdm.sum(axis=0)
```

```python
tdm.sum(axis=1)# 문서별 총 단어 수
```

```python
word_count = pd.DataFrame({'단어':cv.get_feature_names_out(),'빈도':tdm.sum(axis=0).flat})

word_count
```

```python
word_count.sort_values('빈도',ascending=False).head()
```

## 단어 빈도 저장

```python
word_count.to_csv('word_count.csv')
```
