```python
import spacy

nlp = spacy.load('en_core_web_sm')

text = 'Wikipedia is maintained by colunteers.'
doc = nlp(text)
```

```python
for token in doc:
    print(token.text,
         token.lemma_, # 
         token.pos_, #  
         token.tag_, #  
         token.dep_, #   
         token.is_stop)#  
```

```python
def extract_nv(text):
    doc = nlp(text)
    words = []
    for token in doc:
        if token.tag_[0] in 'NV':
            words.append(token.lemma_.lower())
    return words
```

```python
extract_nv('Apple is a company')
```
